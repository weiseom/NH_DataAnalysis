```python
#tips_pct를 구해서 tipdf에 새로운 컬럼명으로 추가하세요 (tip / total_bill * 100)
tipdf['tips_pct'] = tipdf['tip'] / tipdf['total_bill'] * 100
tipdf.head()
```

```python
#성별, 팁비율별로 정렬하여 팁 비율이 높은 정보 3개 출력
tipdf.sort_values(by=['sex','tips_pct'], ascending = False).head(3)
```

```python
#male 정보를 따로 분리해서 크기 정보 조회
tipdf_man = tipdf[tipdf['sex'] == 'Male']
tipdf_man.shape
```

```python
#Female 정보를 따로 분리해서 크기 정보 조회
tipdf_man = tipdf[tipdf['sex'] == 'Female']
tipdf_man.shape
```
