```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
```

```python
y = x * 2
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'g')
```

```python
plt.plot(x,y, label = "A")
plt.plot(x, x**3, label="B")
plt.legend()
```

```python
plt.figure(1,figsize=(20,10))

plt.subplot(2,3,1)
plt.plot(x,y)
plt.subplot(2,3,2)
plt.plot(x,y,'r')
plt.subplot(2,3,3)
plt.plot(x,y,'g')
plt.subplot(2,3,4)
plt.plot(x,y,'y')
plt.subplot(2,3,5)
plt.plot(x,y,'k')
plt.subplot(2,3,6)
plt.plot(x,y,'p')
```

```python
data = {'a': np.arange(50),
       'c': np.random.randint(0,50,50),
       'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data = data, c = 'c', s='d', alpha = 0.5)
```

```python
import pandas as pd

df = pd.DataFrame({'A': list('ABCDABCDAB'),
                  'B': np.random.randint(1,10,10),
                  'C': np.random.randint(10,20,10),
                  'D': np.random.randint(20,30,10)
                  })

df
```

```python
df.plot()

#       pandas  .
```

```python
df.plot(kind='bar')

# kind 
```

# kind
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot
    - 'hexbin' : hexbin plot
